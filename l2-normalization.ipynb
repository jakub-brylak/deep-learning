# L2 normalization (unit normalization)

## There are vectors that we want to normalize

```python
import numpy as np
import matplotlib.pyplot as plt

vectors = np.array([
    [3, 1, 2],
    [1, 4, 2],
    [2, 2, 5]
])

def l2_normalize(v):
    norm = np.linalg.norm(v)
    return v / norm if norm != 0 else v

normalized_vectors = np.array([l2_normalize(v) for v in vectors])
```

## Display comparison before and after vector normalization

```python
def plot_unit_sphere(axes):
    u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:50j]
    x_sphere = np.cos(u) * np.sin(v)
    y_sphere = np.sin(u) * np.sin(v)
    z_sphere = np.cos(v)
    
    sphere1 = axes.plot_surface(
    x_sphere, y_sphere, z_sphere,
    rstride=1, cstride=1, facecolors=plt.cm.viridis((z_sphere + 1) / 2),
    linewidth=0, antialiased=False, alpha=0.1, shade=True
)

def prepareAxes(figure, title, limit, position):
    axes = figure.add_subplot(position, projection='3d')
    axes.set_title(title)
    axes.set_xlim([-limit, limit])
    axes.set_ylim([-limit, limit])
    axes.set_zlim([-limit, limit])
    axes.set_xlabel('X')
    axes.set_ylabel('Y')
    axes.set_zlabel('Z')
    return axes

fig = plt.figure(figsize=(14, 6))

ax1 = prepareAxes(fig, 'Original Vectors', 5, 121)
ax2 = prepareAxes(fig, 'L2-Normalized Vectors', 1.5, 122)

for v in vectors:
    ax1.quiver(0, 0, 0, *v, color='blue')

for v in normalized_vectors:
    ax2.quiver(0, 0, 0, *v, color='red')

# Plot spheres to highlight lengths of the vectors
plot_unit_sphere(ax1)
plot_unit_sphere(ax2)

plt.tight_layout()
plt.show()
```
